This notebook we will be analyzing Bikkes sales data from the bikeSales.csv file.

Loading the packages

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
bike_sales = pd.read_csv('BikeSales.csv')
```

We are only analyzing 1000 rows of the dataset.

```python
bike_sales_V1 = bike_sales.head(1000)
bike_sales_V1
```

```python
plt.figure(figsize=(15,5))
bike_sales_V1['Profit']
plt.plot(bike_sales_V1['Profit'])
```

Profit By Day

```python
profit = bike_sales_V1['Profit']
day = bike_sales_V1['Day']
```

```python
plt.figure(figsize=(15,5))
plt.bar(day, profit)
plt.xlabel('day')
plt.ylabel('profit')
plt.title('Profit by Day')
plt.show()
```

In this chart we can see the profit by days. It look like the highest profit for the day is the 7th day.

Profit last 30 days

```python
plt.figure(figsize=(15,5))
bike_sales_last_10_days = bike_sales.tail(30).reset_index()
plt.plot(bike_sales_last_10_days['Profit'], label = 'Profit')
plt.xticks(range(0,len(bike_sales_last_10_days['Profit'])),bike_sales_last_10_days['Date'], rotation = 90)
plt.xlabel('Date')
plt.ylabel('Profit')
plt.title('Profit last 30 days')
plt.legend()
plt.show()
```

Profit and Revenue last 30 days

```python
plt.figure(figsize=(15,5))
bike_sales_last_10_days = bike_sales.tail(30).reset_index()
plt.plot(bike_sales_last_10_days['Profit'], label = 'Profit')
plt.plot(bike_sales_last_10_days['Revenue'], label = 'Revenue')
plt.xticks(range(0,len(bike_sales_last_10_days['Profit'])),bike_sales_last_10_days['Date'], rotation = 90)
plt.title('Profit and Revenue last 30 days', fontsize = 20)
plt.xlabel('Dates')
plt.legend()
plt.show()
```

Profit and Revenue last 30 days

```python
plt.figure(figsize=(15,5))
bike_sales_last_10_days = bike_sales.tail(30).reset_index()
plt.bar(range(0,len(bike_sales_last_10_days['Profit'])),bike_sales_last_10_days['Profit'], label = 'Profit')
plt.plot(bike_sales_last_10_days['Revenue'], label = 'Revenue', color = 'orange')
plt.xticks(range(0,len(bike_sales_last_10_days['Profit'])),bike_sales_last_10_days['Date'], rotation = 90)
plt.title('Profit and Revenue last 30 days', fontsize = 20)
plt.xlabel('Dates')
plt.legend()
plt.show()
```

Profit and Revenue last 30 days

```python
plt.figure(figsize=(15,5))
bike_sales_last_10_days = bike_sales.tail(30).reset_index()
plt.bar(range(0,len(bike_sales_last_10_days['Revenue'])),bike_sales_last_10_days['Revenue'], label = 'Revenue', color = 'green')
plt.bar(range(0,len(bike_sales_last_10_days['Revenue'])),bike_sales_last_10_days['Cost'], label = 'Cost', color = 'red')
plt.plot(bike_sales_last_10_days['Profit'], label = 'Profit', color = 'blue')
plt.xticks(range(0,len(bike_sales_last_10_days['Profit'])),bike_sales_last_10_days['Date'], rotation = 90)
plt.title('Profit and Revenue last 30 days', fontsize = 20)
plt.xlabel('Dates')
plt.legend()
plt.show()
```

Reveue and Cost last 30 days

```python
weight = .30

bike_sales_last_10_days[['Revenue', 'Cost']].plot(kind='bar', color = ['green', 'red'], figsize =(15,5))
bike_sales_last_10_days['Profit'].plot(color = 'blue', linestyle = '-.', secondary_y = True)

plt.xticks(range(0,len(bike_sales_last_10_days['Profit'])),bike_sales_last_10_days['Date'], rotation = 90)
plt.title('Revenue and Cost last 30 days', fontsize = 20)
plt.legend(loc = 'upper left', fontsize = 14)
plt.xlabel('Dates')
plt.xlim(-weight, len(bike_sales_last_10_days['Profit']) -weight)
plt.show()
```

Revenue and Cost

```python
plt.figure(figsize=(15,5))
plt.scatter(bike_sales_V1['Revenue'], bike_sales_V1['Cost'], color = 'red', marker = '+')
plt.title('Revenue and Cost', fontsize = 20)
plt.legend(loc = 'upper left', fontsize = 14)
plt.xlabel('Revenue')
plt.ylabel('Cost')
plt.show()
```

Relationship between Revenue and Cost by Product

```python
groups = bike_sales_V1[['Product', 'Revenue', 'Cost']].groupby('Product')
groups

fig, ax = plt.subplots(figsize = (15,10))

for Product, group in groups:
    ax.scatter(group['Revenue'], group['Cost'], marker = 'x', label = Product)

plt.title('Relationship between Revenue and Cost By Product', fontsize = 20)
plt.xlabel('Revenue')
plt.ylabel('Cost')
ax.legend()
plt.show()
```

Relationship between Revenue and cost by Product group by month

```python
groups = bike_sales_V1[['Month', 'Revenue', 'Cost']].groupby('Month')
groups

fig, ax = plt.subplots(figsize = (15,10))

for Month, group in groups:
    ax.scatter(group['Revenue'], group['Cost'], marker = 'x', label = Month)

plt.title('Relationship between Revenue and Cost By Product', fontsize = 20)
plt.xlabel('Revenue')
plt.ylabel('Cost')
ax.legend()
plt.show()
```
